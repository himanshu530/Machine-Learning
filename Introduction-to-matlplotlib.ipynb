## Matplotlib example workflow

```python
# Import matplotlib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig,ax = plt.subplots(figsize=(,10))  # Height and width of plot = figsize

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title = "Simple Plot", xlabel="x-axis",ylabel="y_axis")

# 5. Save and show plot
fig.savefig("images_numpy/sample-plot.png")
```

## Two options for subplots



```python
# Subplots option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#Plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(butter_prices.keys(), butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# Plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(butter_prices.keys(), butter_prices.values());
ax[1,1].hist(np.random.randn(1000)); 
#These subplots can be customized
```

## Which method to use.
* when plotting something quickly use pyplot
* when plotting something advanced use Object Oriented API


```python
over_50 = heart_disease[heart_disease["age"]>50]
```

```python
over_50
```

```python
over_50.head()
```

```python
#pyplot method
over_50.plot(kind="scatter",x="age",y="chol",c="target");
# c is used to color
```

```python
# object oriented Method
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
```

```python
# object oriented method from scratch
fig,ax = plt.subplots(figsize=(10,6))

#Plot the data
scatter = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);

#customize the plot
ax.set(title="Heart diseases and cholestrl levels",
      xlabel="age",ylabel="chol");

#Add a legend
ax.legend(*scatter.legend_elements(),title="target");
#Add a horizontal line

ax.axhline(over_50["chol"].mean(),linestyle='--');
fig.savefig("images_numpy/custom-scatter.png")
```

```python
over_50.head()
```

```python
#plotting subplots with common axes age,chol,thalach
fig,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10))

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);

#Customizing ax0
ax0.set(title="Heart disease with different cholestrol levels",xlabel="age",ylabel="chol");
ax0.legend(*scatter.legend_elements(), title="Target");
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");   #Plotting a mean line of cholestrol

#Add data to ax1
scatter1 = ax1.scatter(x = over_50["age"],y=over_50["thalach"], c=over_50["target"]);

# Customizing ax1
ax1.set(title="Heart disease with different heart rate",xlabel="age",ylabel="thalach");
ax1.legend(*scatter1.legend_elements(), title="Target");
ax1.axvline(x=over_50["age"].mean());

#Add title to figure
fig.suptitle("Heart Disease Analysis");
```
